```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
n = int(1e5)
k = 1.5

y_n = np.random.normal(0, 1, n)
y_g = np.random.gamma(1.5, 1, n)

fig, ax = plt.subplots(figsize=(10, 5), ncols=2, nrows=1)
ax[0].set_title('Normal (Gaussian) Distribution')
ax[1].set_title(f'Gamma Distribution (k={k})')

sn.histplot(y_n, ax=ax[0])
sn.histplot(y_g, ax=ax[1])

ax[0].axvline(y_n.mean(), color='red')
ax[1].axvline(y_g.mean(), color='red')

# ax[0].annotate(f'Average: %.3f' % y_n.mean(), (-4, 2400))
# ax[1].annotate(f'Average: %.3f' % y_g.mean(), (6, 2400))

fig.savefig('distribution.png')
```

```python
df_n = pd.DataFrame(y_n, columns=['Value'])
df_n['Distribution'] = 'Normal'
df_g = pd.DataFrame(y_g, columns=['Value'])
df_g['Distribution'] = 'Gamma'
df = pd.concat((df_n, df_g))
g = sn.boxplot(df, y='Value', x='Distribution')
g.figure.savefig('boxplot.png')
```
